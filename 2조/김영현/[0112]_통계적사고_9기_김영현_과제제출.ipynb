```python
import numpy as np
import pandas as pd
```

```python
funding_rounds=pd.read_csv(r"C:\Users\김영현\Desktop\DSL\EDA\crunchbase_1\funding_rounds.csv",index_col=0)
```

```python
funding_rounds.head(4)
```

```python
df_funding_rounds=funding_rounds[['name','raised_amount_usd','post_money_valuation_usd','investor_count','org_uuid','org_name','lead_investor_uuids']]
```

```python
df_funding_rounds.head(4)
```

```python
lst0=[]
for i in df_funding_rounds['name']:
    lst0.append('Debt Financing' in i)

lst1=[]
for i in df_funding_rounds['name']:
    lst1.append('Grant' in i)

lst2=[]
for i in df_funding_rounds['name']:
    lst2.append('Secondary Market' in i)

lst3=[]
for i in df_funding_rounds['name']:
    lst3.append('Convertible Note' in i)
```

```python
debt_firms = df_funding_rounds[lst0 or lst1 or lst2 or lst3]
org_raised_money=  debt_firms.groupby('org_name').sum()[['raised_amount_usd']]
org_raised_money
# 'Debt Financing','Secondary Market','Grant','Convertible Note'를 이용해
#  조달한 자금의 기업별 합
```

```python
org_name = list(org_raised_money.index)
lst=[]
for i in df_funding_rounds['name']:
    lst.append('Series C' in i)
```

```python
firm_series_c = set(df_funding_rounds[lst]['org_name'])
```

```python
lst4=[]
for name in org_name:
    lst4.append(name in firm_series_c)
```

```python
org_raised_money['Series_C']=lst4
```

```python
org_raised_money.head(4)
```

```python
x_train=org_raised_money['raised_amount_usd']
y_train=org_raised_money['Series_C']
```

```python
x_train = (x_train - x_train.mean())/x_train.std()
```

```python
import statsmodels.api as sm
logit = sm.Logit(y_train,x_train).fit()
logit.summary()
```

```python
print(np.exp(logit.params))
```
