```python
import random
```

```python
ipo=fundround_org[fundround_org['status']=='ipo']['raised_amount_usd'].values.tolist()
acquired=fundround_org[fundround_org['status']=='acquired']['raised_amount_usd'].values.tolist()
operating=fundround_org[fundround_org['status']=='operating']['raised_amount_usd'].values.tolist()
closed=fundround_org[fundround_org['status']=='closed']['raised_amount_usd'].values.tolist()
print(len(acquired))
print(len(operating))
print(len(closed))
print(len(ipo))
ipo_ran=random.sample(ipo,100)
acquired_ran=random.sample(acquired,100)
closed_ran=random.sample(closed,100)
operating_ran=random.sample(operating,100)
status_anova=pd.DataFrame({'ipo':ipo_ran,'acquired':acquired_ran,'operating':operating_ran,'closed':closed_ran})
status_anova
```

```python
import seaborn as sns
import scipy.stats as stats
```

```python
sns.kdeplot(data=ipo_ran,color="red",shade=True)
sns.kdeplot(data=operating_ran,color="blue",shade=True)
sns.kdeplot(data=acquired_ran,color="green",shade=True)
sns.kdeplot(data=closed_ran,color="yellow",shade=True)
plt.show()
```

```python
print(stats.shapiro(ipo_ran))
print(stats.shapiro(operating_ran))
print(stats.shapiro(acquired_ran))
print(stats.shapiro(closed_ran))
```

P-value 가 0.05보다 작음 --> 정규분포이라고 할 수 없음. 4개의 status에 따른 amount of fund raised의 평균의 관계를 분석할 때 정규성을 가정할 수 없다.

```python
print('skewness test: {}'.format( round(abs(stats.skew( ipo_ran , bias = False) ),4) ) , sep = '\n')
print('skewness test: {}'.format( round(abs(stats.skew( acquired_ran , bias = False) ),4) ) , sep = '\n')
print('skewness test: {}'.format( round(abs(stats.skew( operating_ran , bias = False) ),4) ) , sep = '\n')
print('skewness test: {}'.format( round(abs(stats.skew( closed_ran , bias = False) ),4) ) , sep = '\n')
```
