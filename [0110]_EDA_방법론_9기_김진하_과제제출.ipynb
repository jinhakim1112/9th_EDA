```python
from scipy import stats
```

```python
t_stat, p_value = stats.ttest_ind(jiwongeum_2014['지원금액(단위_백만원)'],jiwongeum_2022['지원금액(단위_백만원)'], equal_var=True, alternative="two-sided")
print("t-statistics : {}, p-value : {}".format(t_stat, p_value))
```

가장 극단적인 결과를 위해 자료의 가장 첫 시기(2014)와 가장 최근 시기(2022)의 지원금을 비교했을 때, p-value < 0.05이므로 귀무가설을 기각할 수 있고 평균 지원금의 차이는 유의미하다는 결론을 내릴 수 있습니다.

```python
t_stat, p_value = stats.ttest_ind(jiwongeum_2014['지원금액(단위_백만원)'],jiwongeum_2015['지원금액(단위_백만원)'], equal_var=True, alternative="two-sided")
print("t-statistics : {}, p-value : {}".format(t_stat, p_value))
```

극단적인 값을 피하기 위해 2014년도와 바로 다음 년도의 지원금을 비교했을 때, p-value < 0.05이므로 비록 이전 결과보다 p-value가 훨씬 크긴 하지만 귀무가설을 기각할 수 있고 평균 지원금의 차이는 유의미하다는 결론을 내릴 수 있습니다.
